```python
# No tiene d, m, w, Y
# 'ATVEKGGKHKTGPNEKGKKIFVQKCSQCHTVLHGLFGRKTGQA'
class CalculadorRNA():
    
    def codonesDe(self, aa):
        if aa == 'A':
            return ['GCU','GCC','GCA','GCG']
        elif aa == 'R':
            return ['CGU','CGC','CGA','CGG']
        elif aa == 'N':
            return ['AAU','AAC']
        elif aa == 'D':
            return ['GAU','GAC']
        elif aa == 'C':
            return ['UGU','UGC']
        elif aa == 'Q':
            return ['CAA','CAG']
        elif aa == 'E':
            return ['GAA','GAG']
        elif aa == 'G':
            return ['GGU','GGC','GGA','GGG']
        elif aa == 'H':
            return ['CAU','CAC']
        elif aa == 'I':
            return ['AUU','AUC','AUA']
        elif aa == 'L':
            return ['UUA','UUG','CUU','CUC','CUA','CUG']
        elif aa == 'K':
            return ['AAA','AAG']
        elif aa == 'M':
            return ['AUG']
        elif aa == 'F':
            return ['UUU','UUC']
        elif aa == 'P':
            return ['CCU','CCC','CCA','CCG']
        elif aa == 'S':
            return ['UCU','UCC','UCA','UCG']
        elif aa == 'T':
            return ['ACU','ACC','ACA','ACG']
        elif aa == 'W':
            return ['UGG']
        elif aa == 'Y':
            return ['UAU','UAC']
        elif aa == 'V':
            return ['GUU','GUC','GUA','GUG']
        else:
            print('error: el aminoacido no respeta su nombre en el codigo genetico')
            
    def forzarDefinicionARN(self, listCodones):
        rna = ''
        for list in listCodones:
            rna = rna + list[0]
        return rna
    
    def codificarARN(self, peptido):
        rna = []
        for aa in pept:    
            if aa == 'A':
                rna.append(self.codonesDe('A'))
            elif aa == 'R':
                rna.append(self.codonesDe('R'))
            elif aa == 'N':
                rna.append(self.codonesDe('N'))    
            elif aa == 'D':
                rna.append(self.codonesDe('D'))
            elif aa == 'C':
                rna.append(self.codonesDe('C'))    
            elif aa == 'Q':
                rna.append(self.codonesDe('Q'))      
            elif aa == 'E':
                rna.append(self.codonesDe('E'))       
            elif aa == 'G':
                rna.append(self.codonesDe('G'))       
            elif aa == 'H':
                rna.append(self.codonesDe('H'))       
            elif aa == 'I':
                rna.append(self.codonesDe('I'))       
            elif aa == 'L':
                rna.append(self.codonesDe('L'))  
            elif aa == 'K':
                rna.append(self.codonesDe('K'))       
            elif aa == 'M':
                rna.append(self.codonesDe('M'))       
            elif aa == 'F':
                rna.append(self.codonesDe('F'))       
            elif aa == 'P':
                rna.append(self.codonesDe('P'))       
            elif aa == 'S':
                rna.append(self.codonesDe('S'))       
            elif aa == 'T':
                rna.append(self.codonesDe('T'))   
            elif aa == 'W':
                rna.append(self.codonesDe('W'))    
            elif aa == 'Y':
                rna.append(self.codonesDe('Y'))    
            elif aa == 'V':
                rna.append(self.codonesDe('V'))  
            else:
                print('error: uno de los aminoacidos del peptido no respeta el nombre del codigo genetico')
        return rna
        
pept = 'ATVEKGGKHKTGPNEKGKKIFVQKCSQCHTVLHGLFGRKTGQA'
calc = CalculadorRNA()
print(calc.forzarDefinicionARN(calc.codificarARN(pept)))
```
